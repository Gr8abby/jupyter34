```python
import matplotlib.pyplot as plt 

students =[ "Student A","Student B", "Student C", "Student D"]
scores = [98,76,24,68]

colors = ["black","red","green","yellow"]
#creating visualiozations
plt.figure(figsize=(10,10))
plt.bar(students, scores, color=colors)

plt.xlabel("students")
plt.ylabel("Scores")
plt.title("THIS IS A STUDENTS BAR CHART", fontsize=20)

plt.show()
```
