```python
import folium
import pandas as pd

route =[

[9.056288, 7.498658], # Abuja
[8.537010, 4.518735], # Ilorin
[6.520059, 3.387622], # Lagos

]

route_map = folium.Map(location=[8.5, 4.5], zoom_start=6)

for lat, lon, city in zip([9.056288, 8.537010, 6.520059,], [7.498658, 4.518735, 3.387622], ["Abuja", "Ilorin", "Lagos"]):
    folium.Marker([lat, lon], popup=city, icon=folium.Icon(color="blue")).add_to(route_map)



folium.PolyLine(route, color="red", weight=5, opacity=0.8).add_to(route_map)


route_map.save("Delivery_route.html") 
route_map
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sales_data = {
"Region A": {"Product A": 5000, "Product B": 3600, "Product C":7600},
"Region B": {"Product A": 4500, "Product B": 5000, "Product C": 3200},
"Region C":{"Product A": 3000, "Product B": 6000, "Product C": 7600},
"Region D":{"Product A": 2000, "Product B": 4300, "Product C": 5400}
}
df=pd.DataFrame(sales_data)
print(df)

print("\n sum of All product")
sum_product = df.sum(axis=1)
print(sum_product)

print("\n sum of All Region")
sum_region = df.sum(axis=0)
print(sum_region)

#heatmap of sales data

plt.figure(figsize=(8,8))
sns.heatmap(df, annot=True, cmap="coolwarm", fmt="d")
plt.title("Range of Sales Data")
plt.show()
```
